```python
# Import des librairies
import requests
import pandas as pd
import os

# Définition des paramètres de la requête
query = 'journalTitle_s:("Physiological Reviews" OR "Proceedings of the National Academy of Sciences" OR "Nucleic Acids Research" OR "Public Library of Science Biology" OR "Current Biology" OR "British medical journal" OR "The New England Journal of Medicine" OR "Journal of the American Medical Association" OR "Physiological Reviews" OR "Nature" OR "Science" OR "PNAS" OR "eLife" OR "Neuron" OR "Cell" OR Nature* OR "NAR" OR "PLoS Biol" OR "Curr Biol" OR "NEJM" OR "Nature Medicine" OR "Circulation" OR "Blood" OR "Hepatology" OR "Brain" OR "Physiol Rev" OR "Stroke" OR Lancet* OR JAMA* OR BMJ* OR ("JAMA") OR ("Lancet") OR ("BMJ")) AND publicationDateY_i:[2018 TO 2024]'

# Définition des colonnes qui doivent apparaître
fields = 'journalId_i,journalTitle_s,publicationDateY_i,docType_s,structId_i'
rows = 10000  # Nombre de résultats à récupérer par requête
start = 0  # Index de départ des résultats

# URL de base de l'API HAL
base_url = 'https://api.archives-ouvertes.fr/search/INRIA2'

# Initialisation d'une liste pour stocker les résultats
results = []

# Execute des requêtes  jusqu'à ce qu'il n'y ait plus de résultats
while True:
    # Construction de  l'URL de la requête
    url = f'{base_url}?q={query}&fl={fields}&rows={rows}&start={start}'
    print(url)
    
    # Exection de la requête
    response = requests.get(url)
    data = response.json()
    
    # Vérification  des résultats
    docs = data.get('response', {}).get('docs', [])
    if not docs:
        break
    
    # Ajout des résultats à la liste
    results.extend(docs)
    
    # Passeà la page suivante
    start += rows

# Création d' un DataFrame pandas à partir des résultats
df = pd.DataFrame(results)

df.to_json('df.json') #enregistrement en format JSON
df.to_csv('df.csv') #enregistrement en format csv
print('Dataframe créé')
```

```python
# groupement par date afin de compter le nombre d'articles
counts = filtered_df.groupby(['publicationDateY_i', 'journalTitle_s']).size().reset_index(name='count')
counts.head()
```

```python
# Modification du nom des colonnes 
counts.rename(columns={'journalId_i': 'ID', 'publicationDateY_i': 'Annees', 'journalTitle_s': 'Revues', 'count':'Nbre_articles'}, inplace=True)
counts.head(5)
```

```python
# Ajout des domaines (facultatif)
counts['Domaines'] = counts['Revues'].apply(lambda x:'Biology' if x in ['Nature Machine Intelligence','Nature Immunology','Nature Cell Biology','Nature Reviews Neuroscience','Nature Materials', 
                  'Nature Plants', 'Nature Water', 'Nature Ecology & Evolution','Nature Metabolism',                               
                  'Nature Cell Biology', 'Nature Neuroscience', 'Nature Microbiology' ,'Nature Biotechnology' , 'Nature Reviews Physics', 
                  'Nature Reviews Cardiology','Nature Physics','Nature Human Behaviour','Nature Medicine', 'Nature Machine Intelligence ',
                  'Nature Communications','Nature Methods', 'PNAS', 'eLife', 'Neuron', 'Cell','Public Library of Science Biology', 'Current Biology',
                  'NAR', 'PLoS Biol',  'Curr Biol', 'Proceedings of the National Academy of Sciences', 'Nucleic Acids Research',"Nature", "Science"] else   'Medicine')
counts

       
```

```python
table = pd.pivot_table(counts, values='Nbre_articles', index=['Domaines','Revues'],
                       columns=['Annees'], aggfunc="sum")
table.fillna(0, inplace = True)
table['Total'] = table.sum(axis=1)
table = table.astype(int)
table
```
