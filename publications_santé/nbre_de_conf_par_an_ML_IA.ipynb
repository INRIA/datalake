```python
### Création d'un fichier excel avec les formes des Conférences et les formes contrôlées
# et une colonne avec les titres "appauvris" (sans espaces ni caractères spéciaux)

import json
import pandas as pd
import re

# Fonction pour nettoyer le form_title
def clean_title(title):
    # Supprimer les espaces et les caractères spéciaux
    return re.sub(r'\W+', '', title)

# Charger le fichier JSON
with open('valid_conf.json', 'r') as f:
    data = json.load(f)

# Initialiser une liste pour stocker les données
data_list = []

# Parcourir les éléments du fichier JSON
for conference in data:
    title = conference.get('title', '')
    revconf_siid = conference.get('revconf_siid', '')
    
    # Parcourir les formes de chaque conférence
    for form in conference.get('forms', []):
        form_id = form.get('form_id', '')
        form_title = form.get('form_title', '')
        form_year = form.get('form_year', '')
        
        # Nettoyer le form_title
        clean_form_title = clean_title(form_title)
        
        # Ajouter les données à la liste
        data_list.append([form_id, form_title, form_year, title, revconf_siid, clean_form_title])

# Créer un DataFrame pandas à partir des données
df = pd.DataFrame(data_list, columns=['form_id', 'form_title', 'form_year', 'title', 'revconf_siid', 'clean_form_title'])

# Enregistrer le DataFrame dans un fichier Excel
csv_filename = 'conferences_revconf.csv'
df.to_csv(csv_filename, index=False)

print(f"Les données ont été exportées avec succès dans '{csv_filename}'.")

```
